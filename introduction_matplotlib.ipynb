```python
import matplotlib.pyplot as plt
import numpy
plt.subplot(2,2,1)#élément passé en paramètre:      .subplot(line_number(s), column_number(s), position)
x = numpy.arange(1,6)
y_1 = numpy.sin(x)
y_2 = numpy.cos(x)
plt.hist(y_1)
plt.subplot(2,2,4)
plt.scatter(x, y_2, label = "cos(x)")
plt.legend()
plt.show()
```
