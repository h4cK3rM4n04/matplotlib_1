```python
#coding:utf-8
import matplotlib.pyplot as pyp
import numpy as noum
fig, ax = pyp.subplots()
x, y = [x for x in range(1, 6)], [60,240,480,960,1920]
pyp.xlabel("Axe des minutes")
pyp.ylabel("Axe des secondes")
pyp.grid()
pyp.bar(x, y)
#pyp.pie(x) permet de créer un diagramme circulaire
pyp.hist(x)
pyp.xlim(0,6)
pyp.ylim(0,2000)
pyp.scatter(x, y)
pyp.title(pyp.gca())
pyp.gca()
ax.plot(x, y)
ax.set_facecolor("black")
fig.set_facecolor("orange")
pyp.show()
```

```python
#coding:utf-8
import matplotlib.pyplot as plt
import numpy as np
fig, bo = plt.subplots()
x = np.arange(0, 5, 1)
y = np.cos(x)
plt.plot(x, y)
plt.grid()
plt.gca()
plt.scatter(x, y)
fig.set_facecolor("orange")########bleu
bo.set_facecolor("lightblue")
plt.title(plt.gca()) #AxesSubplot etc........
plt.show()
```
