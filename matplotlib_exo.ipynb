**Exercice 2     :** (directive by ChatGPT) 

`Le but de cet exercice est de tracer la courbe de la fonction:` $f(x) = x^3 - 4x^2 + 3x + 1$ `sur l'intervalle` $[-2, 4]$

**1.** Créez une liste de 100 valeurs équidistantes sur l'intervalle $[-2, 4]$ pour $x$.

```python
#coding:utf-8
import matplotlib.pyplot as plt
import numpy
x = numpy.linspace(-2,4,100)#méthode linspace allant de -2 à 4 ayant un len() de 100
```

**2.** Ajoutez une cellule code pour résoudre l'exercice. Commentez chaque étape pour expliquer ce qui se passe.

```python
#définition de la fonction f(x):
def f(x):
    return x**3-4*x**2+3*x+1
#valeurs que prennent l'axe des y
y = f(x)
#tracage de la courbe de f(x) en fonction de x
plt.plot(x,y)
#ajout d'étiquettes pour l'axe x et y
plt.xlabel("axe des x")
plt.ylabel('axe des y')
#ajout d'un titre au graphique
plt.title("courbe de la fonction f(x = x**3-4*x**2+3*x+1)")
#affichage de la courbe
plt.show()
```
